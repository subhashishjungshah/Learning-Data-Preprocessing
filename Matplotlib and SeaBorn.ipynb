```python
x = np.linspace(-5,5,50)
```

```python
# Multiple plot in same graph
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

```python
# Bar Plot
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
languages = ['English','french','Spanish','Nepali']
person = [100,50,140,40]
ax.bar(languages,person)
plt.xlabel('Languages')
plt.ylabel('People')
plt.show()
```

```python
# Pie Chart
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','french','Spanish','Nepali']
person = [100,50,140,40]
ax.pie(person,labels=languages,autopct="%1.1f%%")
plt.show()
```

```python
#Sactter Plot 
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color ='b')
plt.plot()
```

```python
#3d scatter plot
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20* np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```

## Seaborn Datavisualization Library

```python
# importing seaborn library 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Seaborn has some built in datasets 
# total bill vs tip datasets
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# It sets the default theme for our visualaization
sns.set_theme()
```

```python
# visualize this dataset
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# Scatter plot
sns.scatterplot(x='sepal_length', y='petal_length', hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_width', y='petal_width', hue='species',data=iris)
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# Count Plot 
sns.countplot(x = 'class',data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

```python
# barchart
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
# house price dataset
from sklearn.datasets import load_boston
```

```python
house_boston = load_boston()
```

```python
housedf = pd.DataFrame(house_boston.data,columns=house_boston.feature_names)
```

```python
housedf.head()
```

```python
housedf['Price']= house_boston.target
```

```python
housedf.head()
```

```python
# Distribution plot
sns.distplot(housedf['Price'])
```

```python
# this helps to get te distribution of data points
# Correlation matrix Heat Map
corr = housedf.corr()
```

```python
print(corr)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr, cbar=True, square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
